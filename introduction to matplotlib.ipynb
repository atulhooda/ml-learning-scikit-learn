```python
heartdis = pd.read_csv("heart-disease.csv")
heartdis.head()
```

```python
#create a histogran of age
heartdis.plot(x="age",kind="hist")
```

```python
heartdis["age"].plot(kind="hist",bins = 20);
```

```python
heartdis.plot.hist(figsize=(10,30),subplots=True);
```

```python
## Object Oriented Method
```

```python
over_50 = heartdis[heartdis["age"]>50]
over_50.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax.set(title = "heart diesease and cholestrol levels",xlabel = "age",ylabel = "cholestrol");
ax.legend(*scatter.legend_elements(),title="target");
```

```python
## Subplot of chol,age,thalach
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)
scatter = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
ax0.set(title="heart diseases & cholestrol levels",
        ylabel = "cholestrol")
ax0.legend(*scatter.legend_elements(),title = "target")

scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
ax1.set(title="heart diseases & maximum heart rate",
        ylabel = "max heart rate")
ax1.legend(*scatter.legend_elements(),title = "target")
```
